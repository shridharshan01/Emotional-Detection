```python
import cv2 #computer vision
from deepface import DeepFace
from google.colab.patches import cv2_imshow

# Function to analyze emotion and display it on the image
def analyze_and_display_emotion(image_path): #happy.jpg
    # Load the image
    img = cv2.imread(image_path) #img == happy.jpg

    # Check if image was loaded successfully
    if img is None:
        print(f"Error: Could not load image from {image_path}. Please check the file path.")
        return

    # Analyze the image for emotions
    emotion_analysis = DeepFace.analyze(img, actions=['emotion'])

    # Extract the dominant emotion
    dominant_emotion = emotion_analysis[0]['dominant_emotion']
    print(dominant_emotion)
    # Get bounding box coordinates
    # The 'region' dictionary may have more than 4 keys. Access them individually:
    region = emotion_analysis[0]['region']
    x = region['x']
    y = region['y']
    w = region['w']
    h = region['h']

    # Draw rectangle around the face
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Put the emotion text on the image
    cv2.putText(img, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

    # Display the image with emotion
    cv2_imshow(img)

# Path to your image

images=["disgust.jpg","happy.jpg","surprise.jpg","fear.jpg","sad1.jpg","neutral.jpg","anger.jpg"]

for i in images:
  image_path=i
# Call the function
  analyze_and_display_emotion(image_path)
```
